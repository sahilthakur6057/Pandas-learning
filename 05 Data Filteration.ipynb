### LOC AND ILOC

#### The main difference between loc and iloc in pandas is that loc is label-based indexing (using row and column names).
#### While iloc is integer position-based indexing (using row and column numbers).

```python
d = {
    'name':['Sahil','Happy', 'Rohit','Vansh'],
    'age' :[30,20,50,25]
}
```

```python
df = pd.DataFrame(d,index=['A','B','C','D'])
```

```python
df
```

##### FILTERING BY INDEX NAME

```python
df.loc['A']
```

##### FILTERING BY INDEX NUMBER

```python
df.iloc[1]
```

##### INDEXING SLICING  (ROW , COLUMN , VALUE)

```python
df
```

```python
 df.iloc[1:,0]
```

```python
df.iloc[2:,1]
```
